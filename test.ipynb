```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
number_of_line = 10
not_all = 0
all_data = np.array([],dtype=float)
feature_name = ['EulerX','EulerY_standsit','EulerZ_turn','Acc_X','Acc_Y','Acc_Z']
with open('find_segment','r') as f:
    print(f.readline())
    cnt = 0
    for line in f.readlines():
        cnt +=1
        if cnt > number_of_line and not_all == 1:
            break
        line = line.strip().split(',')
        line = np.array(line[2:8]).astype(float)
        all_data = np.append(all_data,line)

```

```python
all_data = np.reshape(all_data,(-1,6))     
#print(all_data[:,0])
#for i in range(len(all_data[0])):
def butter_bandpass(lowcut, highcut,fs,order=5):
    return signal.butter(order, [lowcut, highcut], fs=fs, btype='band')
for i in [1,2]:
    #print(f'{feature_name[i]} th : {all_data[:,i]}')
    tmp_x = list(range(len(all_data[:,i])))
    #input(tmp_x)
    #input(tmp_y)
    a1, b1 = butter_bandpass(5,200,  fs=1080)
    a2, b2 = butter_bandpass(100, 190,  fs=480)
    
    if i == 1:
        filtered = signal.lfilter(a1,b1, all_data[:,i])*2
        plt.ylim(-100,100)
    else:
        filtered = signal.lfilter(a2,b2, all_data[:,i])*1
        plt.ylim(-200,200)
    plt.title(feature_name[i])
    plt.plot(tmp_x, all_data[:,i])
    plt.plot(tmp_x, filtered)
    plt.show()
```
