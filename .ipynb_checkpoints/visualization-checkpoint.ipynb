```python
x = all_data[:,:-1]
y = all_data[:,-1]
x = x.astype(float)
print(y)
```

```python
def inrange(this_range, n):
    if this_range == []:
        return 0
    if n > this_range[0] and n < this_range[1]:
        return 1
def adjust_interval(itrange, itmid):

    if inrange(itrange, 180): # range include 180
        range_1 = [itrange[0], 180]
        range_2 = [-180, -180 + (itrange[1]-180)]
    elif inrange(itrange, -180): # range include 180
        range_1 = [180 - (-180 - itrange[0]), 180]
        range_2 = [-180, itrange[1]]
    else:
        range_1 = itrange
        range_2 = []
    if itmid > 180:
        itmid = -180 + (itmid-180)
    elif itmid < -180:
        itmid = 180 - (-180 - itmid)
    return [range_1,range_2], itmid
        
        
```
